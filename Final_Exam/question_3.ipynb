```python
# You can assume that the function is called with a strictly positive
# integer as first argument and either True or False as second argument,
# if any.
def get_letters(character, number):
    ascii_code = ord(character)
    #     print(ascii_code)

    preceding_ascii_codes = [ascii_code + i for i in range(number)]
    #     print(preceding_ascii_codes)
    for i in range(len(preceding_ascii_codes)):
        if preceding_ascii_codes[i] < 65:
            preceding_ascii_codes[i] += 26
        elif preceding_ascii_codes[i] > 90:
            preceding_ascii_codes[i] -= 26
    #     print(preceding_ascii_codes)


    preceding_characters = [chr(code) for code in preceding_ascii_codes]
    # print(preceding_characters)


    return chr(preceding_ascii_codes[-1]+1), "".join(preceding_characters)

def rhombus(size, shift_right=False):
    '''
    >>> rhombus(1)
    A
    >>> rhombus(1, True)
    A
    >>> rhombus(2)
     BA
    CD
    >>> rhombus(2, True)
    AB
     DC
    >>> rhombus(3)
      CBA
     DEF
    IHG
    >>> rhombus(3, True)
    ABC
     FED
      GHI
    >>> rhombus(4)
       DCBA
      EFGH
     LKJI
    MNOP
    >>> rhombus(4, True)
    ABCD
     HGFE
      IJKL
       PONM
    >>> rhombus(7)
          GFEDCBA
         HIJKLMN
        UTSRQPO
       VWXYZAB
      IHGFEDC
     JKLMNOP
    WVUTSRQ
    >>> rhombus(7, True)
    ABCDEFG
     NMLKJIH
      OPQRSTU
       BAZYXWV
        CDEFGHI
         PONMLKJ
          QRSTUVW
    '''
    char = 'A'
    if shift_right == True:
        reverse = False
    else:
        reverse = True
        
    for i in range(size):
        char, row = get_letters(char, size)
        
        if reverse == True:
            row = row[::-1]
            
        if shift_right == True:
            print(' ' * (i), f'{row}', sep='')
        else:
            print(' ' * (size - (i + 1)), f'{row}', sep='')
        
        if reverse == True:
            reverse = False
        elif reverse == False:
            reverse = True


if __name__ == '__main__':
    import doctest
    doctest.testmod()
```
