```python
# Resetting Sieve
sieve = [True] * (n + 1)

# Using updated notation
for i in range(2, n+1):
    for multiple in range(i*2, n+1, i):
        sieve[multiple] = False

# Printing result
[n for n in range(2, n+1) if sieve[n]]
```

```python
from math import sqrt
n = 14

# Resetting Sieve
sieve = [True] * (n + 1)

# Using updated notation
for i in range(2, round(sqrt(n)+1)):
    if sieve[i] is False:
        # Makes our code more efficient by removing numbers where they are already non primes
        # E.g. we don't need to check multiples of 4 when we have already checked 2
        continue
    for multiple in range(i*2, n+1, i):
        sieve[multiple] = False

# Printing result
[n for n in range(2, n+1) if sieve[n]]
```

```python
def primes_up_to(n):
    sieve = [True] * (n + 1)
    
    for i in range(2, round(sqrt(n)+1)):
        if sieve[i] is False:
            continue
        for multiple in range(i*i, n+1, i):
            sieve[multiple] = False
    
    # Printing result
    # [n for n in range(2, n+1) if sieve[n]]
```

```python
_ = primes_up_to(10_000_000)
```
